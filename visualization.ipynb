```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
energy = np.loadtxt('result.txt', delimiter=',', skiprows=1)
```

```python
times = np.linspace(0, 2.0, 11)
plt.plot(
    times,
    energy,
    label="First order",
    marker="x",
    c="darkmagenta",
    ls="-",
    lw=0.8,
)
plt.ylabel("Energy")
plt.xlabel("Time steps")
plt.suptitle("Observable evolution")
plt.show()
```
